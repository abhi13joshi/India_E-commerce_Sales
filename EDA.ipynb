```python
from datetime import datetime

orders['Order_Date'] = pd.to_datetime(orders['Order_Date'], errors='coerce')

def format_date(date):
    if pd.notnull(date):
        return date.strftime('%d-%m-%Y')
    return ''

# Apply the formatting function to the order_date column
orders['Order_Date'] = orders['Order_Date'].apply(format_date)

orders.head(30)
```

```python
def convert(date_str):
    if date_str:
        parts = date_str.split('-')
        if len(parts) == 3:
            return f"{parts[1]}-{parts[0]}-{parts[2]}"
    return ''

# Apply the conversion function to the order_date column
orders['Order_Date'] = orders['Order_Date'].apply(convert)
orders.head(30)
```

```python
def convert_month(month):
    datetime_obj = datetime.strptime(month, '%b-%y')
    return datetime_obj.strftime('%B-%Y')

targets['Order_Month']=targets['Order_Month'].apply(convert_month)
targets.head()
```

```python
orders.to_csv('data/orders_x.csv', index=False)
targets.to_csv('data/targets_x.csv', index=False)
details.to_csv('data/details_x.csv', index=False)
```
